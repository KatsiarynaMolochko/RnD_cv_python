```python
import random
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
```

```python
width  = 256 #
height = 256
image = np.zeros((width, height, 3), np.uint8)
color = (154, 0, 100)
thickness = -1
center_coordinates = (128, 128)
radius = 128
cv2.circle(image, center_coordinates, radius, color, thickness)
plt.imshow(image)
plt.show()
cv2.imwrite('img.png', image)
```

```python
dot_radius = 1
dot_color = (255, 255, 255)
dot_thickness = -1
dot_center = (0, 0)
dot_amount = 1000
res = [random.randrange(0,256,1) for i in range (dot_amount)]

dot_center = res[0: 1 + 1: 1 ]
dot = 0
for i in range(0, int(dot_amount) - 1):
    dot_center = (res[i], res[i + 1])
    cur_px_color = image[dot_center[0]][dot_center[1]][0]
    # print('color under dot ', cur_px_color)
    # print('coord', dot_center)
    if cur_px_color != 0:
        dot = dot + 1
    cv2.circle(image, dot_center, dot_radius, dot_color, dot_thickness)
plt.imshow(image)
plt.show()
```

```python
print('amount of dots in circle is:', dot)
print('point ratio:', dot / dot_amount)
print('square of circle is:', (height * width * dot / dot_amount) )
print('pi:', (height * width * dot / dot_amount) / radius**2)
```
